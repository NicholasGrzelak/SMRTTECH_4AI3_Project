# Import Needed Libraries

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import fashion_mnist
from keras.models import Sequential,load_model
from keras.layers import InputLayer, Reshape, Conv2DTranspose, LeakyReLU, ReLU
from keras.layers import Dense, Conv2D, MaxPooling2D, Dropout, Flatten, Activation
from keras.utils import plot_model
from keras.optimizers import Adam
import time
```

#Make Custom Functions
**Define Discriminator**

```python
def define_discriminator(in_shape,modname):
	model = Sequential(name=modname)
	model.add(Conv2D(32, (3,3), strides=(2, 2), padding='same', input_shape=in_shape)) #Reduces image to 16,16,32
	model.add(LeakyReLU(alpha=0.2))
	model.add(Conv2D(64, (3,3), strides=(2, 2), padding='same')) #Reduces image to 8,8,64
	model.add(LeakyReLU(alpha=0.2))
	model.add(Flatten()) #Flattens to single vector of 4096
	model.add(Dense(1, activation='sigmoid')) #Final neron chooses if AI generated or not
	# compile model
	opt = Adam(learning_rate=0.0002, beta_1=0.5)
	model.compile(loss='binary_crossentropy', optimizer=opt, metrics=['accuracy'])
	return model
```

**Define Generator**

```python
def define_generator(latent_dim,modname):
	model = Sequential(name=modname)

	n_nodes = 128 * 8 * 8
	model.add(Dense(n_nodes, input_dim=latent_dim)) #Starts with 8192 nodes
	model.add(LeakyReLU(alpha=0.2))
	model.add(Reshape((8, 8, 128)))

	model.add(Conv2DTranspose(64, (3,3), strides=(2,2), padding='same')) #Increases image to 16,16,32
	model.add(LeakyReLU(alpha=0.2))

	model.add(Conv2DTranspose(32, (3,3), strides=(2,2), padding='same')) #Increases image to 32,32,32
	model.add(LeakyReLU(alpha=0.2))

	model.add(Conv2D(3, (3,3), activation='tanh', padding='same')) #Decreases image to 32,32,3
	return model
```

**Define GAN**

```python
def define_gan(generator, discriminator,modname):

	discriminator.trainable = False

	model = Sequential(name=modname)
	model.add(generator)
	model.add(discriminator)

	opt = Adam(learning_rate=0.0002, beta_1=0.5)
	model.compile(loss='binary_crossentropy', optimizer=opt)
	return model
```

**Define Getting Real Samples**

```python
def generate_real_samples(dataset, n_samples):
	ix = np.random.randint(0, dataset.shape[0], n_samples)
	X = dataset[ix]
	y = np.ones((n_samples, 1))
	return X, y
```

**Define Getting AI Samples**

```python
def generate_fake_samples(generator, latent_dim, n_samples):
  x_input = np.random.randn(latent_dim * n_samples)
  x_input = x_input.reshape(n_samples, latent_dim)
  #Normally verbose = 'auto'
  X = generator.predict(x_input,verbose = 0)
  y = np.zeros((n_samples, 1))
  return X, y
```

**Define Saving Plots**

```python
def save_plot(examples, epoch, n=10):
	for i in range(n * n):
		plt.subplot(n, n, 1 + i)
		plt.axis('off')
		plt.imshow(examples[i, :, :, :])
	filename = 'generated_plot_e%03d.png' % (epoch+1)
	plt.savefig(filename)
	plt.close()
```

**Define Evaluator**

```python
def evaluate(epoch, generator, discriminator, dataset, latent_dim, n_samples=100):
	X_real, y_real = generate_real_samples(dataset, n_samples)
	_, acc_real = discriminator.evaluate(X_real, y_real, verbose=0)

	x_fake, y_fake = generate_fake_samples(generator, latent_dim, n_samples)
	_, acc_fake = discriminator.evaluate(x_fake, y_fake, verbose=0)

	print('>Accuracy real: %.0f%%, fake: %.0f%%' % (acc_real*100, acc_fake*100))

	save_plot(x_fake, epoch)
	filename = 'generator_model_%03d.h5' % (epoch + 1)
	generator.save(filename)
```

**Define Loss Writer**

```python
def writetxt(name,txtlist):
  """Writes each entry in a string into a text file"""
  f = open(name +".txt", "w+")
  for i in txtlist: #cycles through each item
    f.write(str(i)+'\n')
  f.close()
```

#Training GAN
**Get Dataset**

```python
#Download the dataset
(x_train,y_train),(x_test,y_test)=tf.keras.datasets.cifar10.load_data()
```

**Visualize Dataset**

```python
#Visualize the Dataset
class_names = ['Airplane','Automobile','Bird','Cat','Deer','Dog','Frog','Horse','Ship','Truck']

plt.figure(figsize=(8,8))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_train[i], cmap=plt.cm.gray)
    plt.xlabel(class_names[int(y_train[i])])
plt.show()
```

**Normalize Data**

```python
#Normalize Data to [0,1]
train_images = x_train / 255.0
test_images = x_test / 255.0
```

**Hyperparameter Initalization**

```python
batch_size = 400 #The number of images per epoch
n_epochs = 1000 #The number of epochs
bat_per_epo = int(train_images.shape[0] / batch_size) 
half_batch = int(batch_size / 2) #size of half batch used to determine amount of AI and real images sent to discriminator
latent_dim = 150 #dimensions of noise sent to Generator
```

**Build Model**

```python
#Build Discriminator
discriminator = define_discriminator((32,32,3),'discriminator')
discriminator.summary()

#Build Generator
generator = define_generator(100,'Generator')
generator.summary()

#Build GAN
gan = define_gan(generator, discriminator,'GAN')
gan.summary()
```

**Train Model**

```python
dlossavg = [] #saves all the average discriminator loss
glossavg = [] #saves all the average generator loss

start_time = time.time() #Starts time to get how long training

for i in range(n_epochs):
    print("Epoch: " + str(i)) #Prints out current Epoch
    totaldloss = 0 #sets the total discriminator loss in epoch
    totalgloss = 0 #sets the total generator loss in epoch
    for j in range(bat_per_epo):

      X_real, y_real = generate_real_samples(train_images, half_batch) #Gets real images and labels them real
      X_fake, y_fake = generate_fake_samples(generator, latent_dim, half_batch) #Gets AI images and labels them as AI
      X, y = np.vstack((X_real, X_fake)), np.vstack((y_real, y_fake)) #Mixes AI and real images into 1 dataset
      d_loss, _ = discriminator.train_on_batch(X, y) #gets the discriminator loss by training

      X_gan = np.random.randn(latent_dim * batch_size) #Makes normal distribution of size latent_dim * batch_size
      X_gan = X_gan.reshape(batch_size, latent_dim) #Reshapes to proper size
      y_gan = np.ones((batch_size, 1)) #fills the batch with ones, signifies AI generated
      g_loss = gan.train_on_batch(X_gan, y_gan) #Trains generator and gets loss

      #print('>%d, %d/%d, d=%.3f, g=%.3f' % (i+1, j+1, bat_per_epo, d_loss, g_loss))

      totaldloss = totaldloss + d_loss #adds together all loss of discriminator in epoch
      totalgloss = totalgloss + g_loss #adds together all loss of generator in epoch

    totaldloss = totaldloss / (bat_per_epo-1) #After all batches completed finds average discriminator loss
    totalgloss = totalgloss / (bat_per_epo-1) #After all batches completed finds average generator loss

    dlossavg.append(totaldloss) #Saves average discriminator loss
    glossavg.append(totalgloss) #Saves average generator loss

    if (i+1) % 10 == 0: #if divisible by 10 saves the model
      evaluate(i, generator, discriminator, train_images, latent_dim) #saves image set with generator model
      writetxt("Gen_loss",glossavg) #Writes average generator loss
      writetxt("Dis_loss",dlossavg) #Writes average discriminator loss

end_time = time.time() #Ends the training time

writetxt('Gen_loss',glossavg) #Makes a final version of generator losses if training stops
writetxt('Dis_loss',dlossavg) #Makes a final version of discriminator losses if training stops

print('Training Time', end_time-start_time) #Prints training time
print('First Values', dlossavg[0], glossavg[0]) #prints first epoch average losses
print('Last Values', dlossavg[-1], glossavg[-1]) #prints last epoch average losses
```
